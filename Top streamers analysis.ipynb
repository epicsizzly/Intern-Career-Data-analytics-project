```python
import pandas as pd


df = pd.read_csv("youtubers_df.csv")
```

```python
df.head(15)
```

# Data exploration

```python
print("\nDataset structure:")
print(df.info())
```

```python
#missing data
print("\nMissing data:")
print(df.isnull().sum())
```

```python

# Summary statistics to identify outliers
print("\nSummary statistics:")
print(df.describe())
```

```python
# Drop rows with missing values in the "Categories" column
df = df.dropna(subset=['Categories'])

# Fill missing values in the "Categories" column with "Unknown"
df['Categories'].fillna('Unknown', inplace=True)
```

```python
print("Remaining missing data:")
print(df.isnull().sum())
```

```python
df.head(15)
```

# Trend analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Group by Categories and count the number of streamers in each category
category_counts = df['Categories'].value_counts()
category_counts
```

```python
# Plot the top categories
plt.figure(figsize=(10, 6))
category_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular Categories')
plt.xlabel('Category')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the correlation coefficients between subscribers, likes, and comments
correlation_matrix = df[['Suscribers', 'Likes', 'Comments']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### So, there's a tendancy that if subscribers increases so likes and comments increases, so for likes and comments because there's a positive correlation between them

# Audience study

```python
# Group by Country and count the number of streamers in each country
country_counts = df['Country'].value_counts()
country_counts
```

```python
# Plot the distribution of streamers' audiences by country
top_10_countries = country_counts.head(10).index

plt.figure(figsize=(12, 8))
sns.countplot(x='Country', hue='Categories', data=df[df['Country'].isin(top_10_countries)], palette='viridis')
plt.title('Distribution of Content Categories by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45)
plt.legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Yes, there's regional references. Every Country has its different taste of Catrgory.

# Perform metrics

```python
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()

# Visualize average metrics with annotations
plt.figure(figsize=(10, 6))
ax = average_metrics.plot(kind='bar', color='skyblue')

# Add annotations to each bar
for i, v in enumerate(average_metrics):
    ax.text(i, v + 0.5, f'{v:.2f}', ha='center', va='bottom')

plt.title('Average Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Check for anomalies or outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Suscribers', 'Visits', 'Likes', 'Comments']], palette='Set2')
plt.title('Distribution of Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

### we can see there's such a difference of outliers in each of the category

## Content Categories

```python
# Count streamers by category
category_counts = df['Categories'].value_counts()

# Visualize category distribution
plt.figure(figsize=(12, 8))
category_counts.head(10).plot(kind='bar', color='lightgreen')
plt.title('Distribution of Streamers by Content Category (Top 10)')
plt.xlabel('Content Category')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze performance metrics by category
average_metrics_by_category = df.groupby('Categories')[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()

# Visualize performance metrics by category
plt.figure(figsize=(12, 8))
average_metrics_by_category.mean(axis=1).sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Average Performance Metrics by Content Category')
plt.xlabel('Content Category')
plt.ylabel('Average Value')
plt.xticks(rotation=90)
plt.show()
```

### Yes, the Videojuegos, Humor Channels has exceptional perfomance metrics!

##  Brands and Collaborations

```python
import seaborn as sns

# Define high-performance streamers (e.g., top 10% based on performance metrics)
high_performance_criteria = df[['Suscribers', 'Visits', 'Likes', 'Comments']].quantile(0.9)

# Identify high-performance streamers
high_performance_streamers = df[(df['Suscribers'] >= high_performance_criteria['Suscribers']) &
                                (df['Visits'] >= high_performance_criteria['Visits']) &
                                (df['Likes'] >= high_performance_criteria['Likes']) &
                                (df['Comments'] >= high_performance_criteria['Comments'])]
```

```python
# Gather data on brand collaborations (not implemented, requires additional data)

# Analyze correlation between performance metrics and brand collaborations
correlation_metrics = high_performance_streamers[['Suscribers', 'Visits', 'Likes', 'Comments']].corr()

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_metrics, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Performance Metrics vs. Brand Collaborations')
plt.show()
```

### So, The high correlation coefficient suggests an exceptionally strong positive relationship betweeneach category, indicating that increased one value correspond closely to higher other value counts among the highest performing streamers analyzed.

## Benchmarking

```python
# Calculate average performance metrics
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()

# Identify streamers above average
above_average_streamers = df[(df['Suscribers'] > average_metrics['Suscribers']) &
                             (df['Visits'] > average_metrics['Visits']) &
                             (df['Likes'] > average_metrics['Likes']) &
                             (df['Comments'] > average_metrics['Comments'])]

# Rank streamers by performance
ranked_streamers = above_average_streamers[['Username', 'Suscribers', 'Visits', 'Likes', 'Comments']].sort_values(by=['Suscribers', 'Visits', 'Likes', 'Comments'], ascending=False)

# Print top-performing content creators
top_performers = ranked_streamers.head(10)
print("Top Performing Content Creators:")
print(top_performers)
```
